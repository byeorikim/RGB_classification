# Data Augmentation
- Data augmentation through segmentation

## 1. import library

```python
import numpy as np
import matplotlib.image as img
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.colors as colors
from skimage import color
from skimage import io
import os
from PIL import Image
```

## 2. load data

### 2.1 load red image

```python
red_image_list = []
for i in os.listdir('./red_image/'):
    if 'jpeg' in i:
        red_image_list.append(np.array(Image.open('./red_image/%s'%i)))
    
# check
print(len(red_image_list))
print(red_image_list[0].shape)
plt.imshow(red_image_list[0])
```

### 2.2 load blue image

```python
blue_image_list = []
for i in os.listdir('./blue_image/'):
    if 'jpeg' in i:
        blue_image_list.append(np.array(Image.open('./blue_image/%s'%i)))
    
# check
print(len(blue_image_list))
print(blue_image_list[0].shape)
plt.imshow(blue_image_list[0])
```

### 2.3 load green image

```python
green_image_list = []
for i in os.listdir('./green_image/'):
    if 'jpeg' in i:
        green_image_list.append(np.array(Image.open('./green_image/%s'%i)))
    
# check
print(len(green_image_list))
print(green_image_list[0].shape)
plt.imshow(green_image_list[0])
```

## 4. To split and save an image
- collect 256 x 256 size data

### 4.1 red data

```python
count = 1
for i in red_image_list:
    for j in range(11):
        for k in range(11):
            img = i[256*(j):256*(j+1), 256*k:256*(k+1), :]
            plt.imsave('red_data/red_data%d.jpeg'%count, img)
            count = count+1
```

```python
print(count)
```

### 4.2 blue data

```python
count = 1
for i in blue_image_list:
    for j in range(11):
        for k in range(11):
            img = i[256*(j):256*(j+1), 256*k:256*(k+1), :]
            plt.imsave('blue_data/blue_data%d.jpeg'%count, img)
            count = count+1
            
print(count)
```

### 4.3 green data

```python
count = 1
for i in green_image_list:
    for j in range(11):
        for k in range(11):
            img = i[256*(j):256*(j+1), 256*k:256*(k+1), :]
            plt.imsave('green_data/green_data%d.jpeg'%count, img)
            count = count+1
            
print(count)
```
